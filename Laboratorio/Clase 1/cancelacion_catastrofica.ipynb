## Cancelación catastrófica

```python
import numpy as np

xs = []
ys1 = []
ys2 = []
for x in np.arange(-4e-8,4e-8,1e-9):
    xs.append(x)
    y1 = (1.0 - np.cos(x))/x**2.0
    y2 = (2.0*np.sin(x/2.0)*np.sin(x/2.0))/x**2.0
    ys1.append(y1)
    ys2.append(y2)
    
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(xs,ys1,'b.-')
plt.plot(xs,ys2,'r.-')
plt.grid()
plt.legend(['1-cos','2sin^2'])

```
