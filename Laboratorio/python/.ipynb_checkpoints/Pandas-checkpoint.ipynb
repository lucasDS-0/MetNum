Importar Pandas

```python
import pandas as pd
```

Leer el dataset desde: https://data.boston.gov/dataset/6220d948-eae2-4e4b-8723-2dc8e67722a3/resource/f4495ee9-c42c-4019-82c1-d067f07e45d2/download/tmp7_f32p54.csv

```python
df_dataset = pd.read_csv('https://data.boston.gov/dataset/6220d948-eae2-4e4b-8723-2dc8e67722a3/resource/f4495ee9-c42c-4019-82c1-d067f07e45d2/download/tmp7_f32p54.csv', encoding ="utf-8")
```

Mostrar las primeras 10 filas

```python
df_dataset[:10]
```

Mostrar las columnas del dataframe

```python
df_dataset.columns
```

¿Cuántas filas tiene el dataset?

```python
len(df_dataset)
```

¿Cuál es el código de ofensa más realizada?

```python
df_dataset['OFFENSE_CODE'].value_counts().index[0]
```

¿Cuál es el código de ofensa más realizada los días miércoles?

```python
df_dataset[df_dataset['DAY_OF_WEEK'] == 'Wednesday']['OFFENSE_CODE'].value_counts().index[0]
```

Ordenar el dataframe por el campo Street y mostrar las 10 primeras entradas

```python
df_dataset.sort_values(by=['STREET'])[:10]
```

Pasar a minúsculas todos las descripciones de ofensas

```python
df_dataset['OFFENSE_DESCRIPTION'] = df_dataset.apply(lambda row: row['OFFENSE_DESCRIPTION'].lower(),axis =1)
df_dataset
```

¿Cuál es la distribución en deciles de la Latitud de los crímenes?

```python
pd.qcut(df_dataset['Lat'],q=10, retbins=True)
```

Crear una nueva columna con la longitud de la calle donde sucedió el crimen

```python
df_dataset = df_dataset.assign(long2=df_dataset['Long'].values)
df_dataset
```

¿Cuál es la suma de las longitudes calculadas?

```python
df_dataset['Long'].sum()
```

¿Y la mediana?

```python
df_dataset['Long'].median()
```

¿Cuál es el promedio de esas mismas longitudes pero agrupando los datos en cada día de la semana?

```python
df_dataset.groupby(['DAY_OF_WEEK'])['Long'].mean()
```
